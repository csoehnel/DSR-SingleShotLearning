```python
from keras import backend as K
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read and plot image using mpimg (image path: data_aug/meerkat.jpg)

# add your implementation
img = mpimg.imread("./data_aug/meerkat.jpg")
plt.imshow(img)
```

```python
# use ImageDataGenerator from keras.preprocessing.image to generate different variations of the above image

# add your implementation
datagen = ImageDataGenerator(
    rescale=1./255,
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    shear_range=0.2,
    zoom_range=0.2)
```

```python
from ipywidgets import interact

@interact(x = (0, 100))
def randomMeerkat(x):
    plt.imshow(datagen.random_transform(img))
```
